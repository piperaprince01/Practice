```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
X = df_vehicle.drop('class',axis=1)
Y = df_vehicle['class']
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=.3,random_state=1)
```

```python
X_centered = X - X.mean(axis=0)
```

```python
pca = PCA(n_components=2)
pca.fit(X_centered)
```

```python
X_pca = pca.transform(X_centered)
```

```python
X_pca.shape
```

```python
from sklearn.cluster import KMeans
```

```python
k = [] #Track number of clusters
sse = [] #Track SSE
```

```python
#Run it from 2 to 20
for i in range(2,21):    
    new_model = KMeans(n_clusters=i)
    new_model.fit(X_train)
    k.append(i)
    sse.append(new_model.inertia_)
    print ('Cluster(s):' + str(i) + ' SSE: ' + str(new_model.inertia_))
```

```python
sse
```

```python
import matplotlib.pyplot as plt

plt.scatter(k, sse)
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
svm = SVC()
parameters = {'kernel':('linear', 'rbf'), 'C':(0.01, 0.05, 0.5, 1),'gamma': (1,2,3,'auto'),'decision_function_shape':('ovo','ovr'),'shrinking':(True,False)}
clf = GridSearchCV(svm, parameters)
clf.fit(X_train,y_train)
print("accuracy:"+str(np.average(cross_val_score(clf, X_train, y_train, scoring='accuracy'))))
```
